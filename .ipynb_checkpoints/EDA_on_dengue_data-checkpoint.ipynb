## Creating a comprehensive analysis with multiple charts

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('Dengue Cases Analysis - Hyderabad', fontsize=16, fontweight='bold')

# 1. Age distribution
axes[0,0].hist(df['Age'], bins=20, color='skyblue', alpha=0.7, edgecolor='black')
axes[0,0].set_title('Age Distribution of Dengue Cases')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Number of Cases')

# 2. Gender distribution
gender_counts = df['Gender'].value_counts()
axes[0,1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue'])
axes[0,1].set_title('Gender Distribution')

# 3. Outcome distribution
outcome_counts = df['Outcome'].value_counts()
axes[0,2].bar(outcome_counts.index, outcome_counts.values, color=['green', 'orange', 'red', 'darkred'])
axes[0,2].set_title('Treatment Outcomes')
axes[0,2].set_xlabel('Outcome')
axes[0,2].set_ylabel('Number of Cases')
axes[0,2].tick_params(axis='x', rotation=45)

# 4. Cases by year
yearly_cases = df['Year'].value_counts().sort_index()
axes[1,0].plot(yearly_cases.index, yearly_cases.values, marker='o', linewidth=2, markersize=6)
axes[1,0].set_title('Dengue Cases by Year')
axes[1,0].set_xlabel('Year')
axes[1,0].tick_params(axis="x",rotation=45)
axes[1,0].set_ylabel('Number of Cases')
axes[1,0].grid(True, alpha=0.3)

# 5. Platelet count distribution by outcome
df_clean = df.dropna(subset=['Platelet_Count'])
axes[1,1].boxplot([df_clean[df_clean['Outcome'] == outcome]['Platelet_Count'] for outcome in df_clean['Outcome'].unique()], 
                  labels=df_clean['Outcome'].unique())
axes[1,1].set_title('Platelet Count by Outcome')
axes[1,1].set_xlabel('Outcome')
axes[1,1].set_ylabel('Platelet Count')
axes[1,1].tick_params(axis='x', rotation=45)

# 6. Top wards with most cases
ward_counts = df['Ward'].value_counts().head(10)
axes[1,2].barh(range(len(ward_counts)), ward_counts.values)
axes[1,2].set_yticks(range(len(ward_counts)))
axes[1,2].set_yticklabels(ward_counts.index)
axes[1,2].set_title('Top 10 Wards by Cases')
axes[1,2].set_xlabel('Number of Cases')

plt.tight_layout()
plt.show()
```

Age & Platelet_Count: These two variables show almost no linear correlation with each other or with any of the environmental factors (Temperature, Humidity, Rainfall).

Environmental Factors & Platelet_Count: 'Temperature', 'Humidity', and 'Rainfall' have negligible linear impact on 'Platelet_Count' in this dataset.

```python
# Age vs Platelet Count colored by Outcome


# Use seaborn's colorblind-friendly palette
palette = sns.color_palette("colorblind", 4)  # 4 distinct, friendly colors

# Assign improved colors to each outcome
outcome_colors = {
    'Recovered': palette[0],      # Blue
    'Hospitalized': palette[1],   # Orange
    'Critical': palette[2],       # Green
    'Dead': palette[3]            # Red-ish
}

plt.figure(figsize=(10, 6))
for outcome in df['Outcome'].unique():
    subset = df[df['Outcome'] == outcome]
    plt.scatter(subset['Age'], subset['Platelet_Count'], 
                label=outcome, alpha=0.7, edgecolors='w', s=60,
                color=outcome_colors.get(outcome, 'gray'))

plt.title('Age vs Platelet Count by Outcome', fontsize=14, fontweight='bold')
plt.xlabel('Age')
plt.ylabel('Platelet Count')
plt.legend(title='Outcome')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Total cases by ward
ward_cases = df['Ward'].value_counts()
```

```python
#Top 10 High Risk Wards by Total Cases
top_wards = ward_cases.head(10)

plt.figure(figsize=(10, 6))
plt.barh(range(len(top_wards)), top_wards.values, color='darkred', alpha=0.7)
plt.yticks(range(len(top_wards)), top_wards.index)
plt.xlabel('Number of Cases')
plt.title('Top 10 High Risk Wards by Total Cases', fontweight='bold')
plt.grid(axis='x', alpha=0.3)


for i, v in enumerate(top_wards.values):
    plt.text(v + 5, i, str(v), va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Ward severity analysis
ward_severity = df.groupby('Ward')['Outcome'].apply(
    lambda x: ((x == 'Critical').sum() + (x == 'Dead').sum()) / len(x) * 100).sort_values(ascending=False)
```

```python
ward_severity
```

```python
# Combined risk score (cases + severity)
ward_risk_score = pd.DataFrame({
    'Cases': ward_cases,
    'Severity_Rate': ward_severity,
    'Cases_Normalized': (ward_cases - ward_cases.min()) / (ward_cases.max() - ward_cases.min()) * 100,
    'Risk_Score': ((ward_cases - ward_cases.min()) / (ward_cases.max() - ward_cases.min()) * 0.7 + 
                   ward_severity / 100 * 0.3) * 100
}).sort_values('Risk_Score', ascending=False)
```

```python
# Ward Risk Score (Combined metric)
top_risk_wards = ward_risk_score.head(10)
colors = plt.cm.Reds(np.linspace(0.4, 0.9, len(top_risk_wards)))

plt.figure(figsize=(10, 6))
plt.barh(range(len(top_risk_wards)), top_risk_wards['Risk_Score'],color=colors)
plt.yticks(range(len(top_risk_wards)), top_risk_wards.index)
plt.xlabel('Risk Score')
plt.title('Top 10 Wards by Combined Risk Score\n(70% Cases + 30% Severity)', fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# Add risk scores on bars
for i, v in enumerate(top_risk_wards['Risk_Score']):
    plt.text(v + 1, i, f'{v:.1f}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Monthly rainfall vs cases

# Filter data with rainfall information
rainfall_data = df[df['Rainfall'].notna()].copy()
print(f"Analyzing {len(rainfall_data)} cases with rainfall data ({len(rainfall_data)/len(df)*100:.1f}% of total)")

rainfall_data['Rainfall_Category'] = pd.cut(rainfall_data['Rainfall'], 
                                          bins=[0, 1, 10, 50, float('inf')], 
                                          labels=['No Rain (0-1mm)', 'Light (1-10mm)', 'Moderate (10-50mm)', 'Heavy (50mm+)'],
                                          include_lowest=True)


monthly_rainfall = rainfall_data.groupby(rainfall_data['Date_of_diagnosis'].dt.month).agg({
    'Rainfall': ['mean', 'sum', 'count'],
    'Patient_ID': 'count'
}).round(2)

monthly_rainfall.columns = ['Avg_Rainfall', 'Total_Rainfall', 'Days_with_Rain', 'Cases']
monthly_rainfall['Month_Name'] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                                  'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
cases_by_month = [monthly_rainfall.loc[i, 'Cases'] if i in monthly_rainfall.index else 0 for i in range(1, 13)]
rainfall_by_month = [monthly_rainfall.loc[i, 'Avg_Rainfall'] if i in monthly_rainfall.index else 0 for i in range(1, 13)]

fig, ax = plt.subplots(figsize=(10, 6))
ax2 = ax.twinx()
ax.bar(months, cases_by_month, alpha=0.7, color='steelblue', label='Cases')
ax2.plot(months, rainfall_by_month, color='darkgreen', marker='o', linewidth=3, markersize=8, label='Avg Rainfall (mm)')

ax.set_ylabel('Number of Cases', color='steelblue', fontweight='bold')
ax.set_title('Monthly Cases vs Average Rainfall', fontweight='bold')
ax.tick_params(axis='y', labelcolor='steelblue')
ax2.set_ylabel('Average Rainfall (mm)', color='darkgreen', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='darkgreen')
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
#Rainfall Categories vs Cases

rainfall_category_analysis = rainfall_data.groupby('Rainfall_Category').agg({
    'Patient_ID': 'count',
    'Outcome': lambda x: ((x == 'Critical').sum() + (x == 'Dead').sum()) / len(x) * 100
}).round(2)
rainfall_category_analysis.columns = ['Cases', 'Severity_Rate']


rainfall_cats = ['No Rain\n(0-1mm)', 'Light\n(1-10mm)', 'Moderate\n(10-50mm)', 'Heavy\n(50mm+)']
cases_by_rain = rainfall_category_analysis['Cases'].values
severity_by_rain = rainfall_category_analysis['Severity_Rate'].values

fig, ax = plt.subplots(figsize=(10, 6))
ax2 = ax.twinx()
ax.bar(rainfall_cats, cases_by_rain, alpha=0.7, color='lightblue', label='Cases')
ax2.plot(rainfall_cats, severity_by_rain, color='red', marker='s', linewidth=3, markersize=8, label='Severity %')

ax.set_ylabel('Number of Cases', color='steelblue', fontweight='bold')
ax.set_title('Cases and Severity by Rainfall Intensity', fontweight='bold')
ax.tick_params(axis='y', labelcolor='steelblue')
ax2.set_ylabel('Severity Rate (%)', color='red', fontweight='bold')
ax2.tick_params(axis='y', labelcolor='red')
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

From this chart we have understood that severity rate is more when the rainfall intensity was increasing.

**===== Top Key Insights from the Pair Plot (Outcome-wise Analysis): =====**

**Platelet Count is the Most Critical Outcome Predictor**

Patients with low platelet counts are strongly associated with Critical and Dead outcomes, while higher counts are seen in **Recovered** and **Hospitalized** cases.

**Environmental Factors Have Little Impact on Outcome Severity**

Temperature, Humidity, and Rainfall show no strong visual or statistical separation between different outcomes, indicating they don't directly influence how severe a case becomes.

**Slight Age-related Risk for Severe Outcomes**

Older patients have a slightly higher tendency to result in Dead outcomes, though age distributions for different outcomes still largely overlap.

**Temperature and Humidity are Negatively Correlated**

As expected, there's a clear inverse relationship: higher temperatures generally come with lower humidity, though this does not affect the outcome directly.

**Most Variables Show Overlapping Outcome Distributions**

Apart from Platelet Count, most other features do not provide clear separation between outcomes, making multi-variable or model-based analysis essential for better prediction.
