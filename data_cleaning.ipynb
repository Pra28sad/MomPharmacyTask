## Data Cleaning 

```python
df.isnull().sum()
```

```python
df["Date_of_diagnosis"].dtype   ## this is object data type so i need to convert it into date type column
```

```python
df["Date_of_diagnosis"] = pd.to_datetime(df["Date_of_diagnosis"])
```

```python
df["Date_of_diagnosis"].dtype
```

```python
df["Month"] = df["Date_of_diagnosis"].dt.month
df["Year"] = df["Date_of_diagnosis"].dt.year
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df = df[df["Year"] != 2025]
```

```python
df.shape
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df["Age_group"] = pd.cut(df["Age"],bins=[0,18,35,60,100],labels=["Child","Young Adult","Adult","Senior"])

```
